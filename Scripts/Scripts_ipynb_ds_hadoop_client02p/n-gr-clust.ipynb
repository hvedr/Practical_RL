```python
def ngr(s, n, res = dict(), prefix = ''):
    for i in range(len(s)-n+1):
        k = prefix + s[i:(i+n)]
        res[k] = res[k] + 1 if k in res else 1
    return res
```

```python
ngr("привет", 1, dict(), '')
```

```python
def tngr(talk, n):
    res = dict()
    for s in talk:
        if s[:3] == 'A: ':
            ngr(s[3:], n, res, 'a_')
        elif s[:3] == 'O: ':
            ngr(s[3:], n, res, 'o_')
    return res
    
```

```python
def cos(d1, d2):
    norm = sum([d1[k]*d1[k] for k in d1])*sum([d2[k]*d2[k] for k in d2])
    norm = norm**0.5
    if norm == 0:
        return 0
    return sum([d1[k]*d2[k] for k in d1 if k in d2])/norm
```

```python
n = 3
ngrs = {k: tngr(res[k], n) for k in res}
```
