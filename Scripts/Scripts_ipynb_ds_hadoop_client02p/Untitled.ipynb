```python
def f(x,y):
    return x+y

def phi(x,y):
    return x**2 + y**2 - 1

def L(x,y,l):
    return f(x,y) + l**2 * phi(x,y)
x,y,l = 0.5,0.2,0.1
step = 0.2
d = 0.01
thres = 1e-2

#Gradientdescent
for _ in range(100):
    x,y,l = x if x!=0 else x + 1e+10, y if y!=0 else y + 1e+10, l if l!=0 else l + 1e+10,
    L_prev = L(x,y,l)
    L_x = (L(x*(1+d),y,l) - L(x,y,l)) / (x*d)
    L_y = (L(x,y*(1+d),l) - L(x,y,l)) / (y*d)
    L_l = (L(x,y,l*(1+d)) - L(x,y,l)) / (l*d)
    x -= L_x * step
    y -= L_y * step
    l -= L_l * step
    if(L(x,y,l) > L_prev - thres): step *= 0.5
    print('{} {} {} {}'.format(L(x,y,l),x,y,l))
```
