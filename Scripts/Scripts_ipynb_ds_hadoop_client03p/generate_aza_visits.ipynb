```python
import datetime
init_query = '''
CREATE TABLE `user_kposminin.ccall_visits`(
  `phone_mobile` varchar(35), 
  `approve` int, 
  `call_ymd` string, 
  `id` string, 
  `urlfr` string, 
  `cnt` int, 
  `duration` int, 
  `avg_hour` int)
      partitioned by (`ymd` string)
ROW FORMAT SERDE
  'org.apache.hadoop.hive.serde2.columnar.ColumnarSerDe'
STORED AS INPUTFORMAT
  'org.apache.hadoop.hive.ql.io.RCFileInputFormat'
OUTPUTFORMAT
  'org.apache.hadoop.hive.ql.io.RCFileOutputFormat'
'''


gen_visits_for_d_day_query = '''

insert overwrite table user_kposminin.ccall_visits partition (ymd)
select 
  i.phone_mobile,  
  i.approve,
  i.ymd as call_ymd,
  v.id,
  v.urlfr, 
  v.cnt, 
  v.duration, 
  v.avg_hour, 
  v.ymd
from prod_features_liveinternet.visits v
inner join user_kposminin.aza_liru_trig_train_id i on i.id = v.id
where v.ymd = '#ymd'
'''


start_day = datetime.datetime(2016,1,1).date()

print(';'.join([init_query] + [gen_visits_for_d_day_query.replace('#ymd',(start_day + datetime.timedelta(days=i)).strftime('%Y-%m-%d')) for i in range(365)]))
```
